Q5: Curse of Dimensionality and PCA

The curse of dimensionality refers to the problem that arises when the number of features or dimensions in a dataset is very large leading to increased complexity, decreased model performance adn an increased risk of overfitting.

In this dataset, we have 5 features: bill_length_mm, bill_depth_mm, flipper_length_mm, body_mass_g, and sex. While this is not an extremely high-dimensional dataset, we can still perform Principal Component Analysis (PCA) to reduce the dimensionality and visualize the data more effectively.

```python
import pandas as pd
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

df_numeric = df[['bill_length_mm', 'bill_depth_mm', 'flipper_length_mm', 'body_mass_g']]

# Perform PCA
pca = PCA(n_components=2)
pca_df = pca.fit_transform(df_numeric)

# Create a new DataFrame with the PCA components
pca_df = pd.DataFrame(pca_df, columns=['PC1', 'PC2'])

# Plot the PCA components
plt.figure(figsize=(10, 6))
plt.scatter(pca_df['PC1'], pca_df['PC2'])
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA Components')
plt.show()
```

Q6: Visualizing bill_length_mm vs bill_depth_mm

```python
import matplotlib.pyplot as plt

# Scatter Plot
plt.figure(figsize=(10, 6))
plt.scatter(df['bill_length_mm'], df['bill_depth_mm'])
plt.xlabel('Bill Length (mm)')
plt.ylabel('Bill Depth (mm)')
plt.title('Scatter Plot')
plt.show()

# Line Plot
plt.figure(figsize=(10, 6))
plt.plot(df['bill_length_mm'], df['bill_depth_mm'])
plt.xlabel('Bill Length (mm)')
plt.ylabel('Bill Depth (mm)')
plt.title('Line Plot')
plt.show()

# Histogram
plt.figure(figsize=(10, 6))
plt.hist2d(df['bill_length_mm'], df['bill_depth_mm'], bins=20)
plt.xlabel('Bill Length (mm)')
plt.ylabel('Bill Depth (mm)')
plt.title('Histogram')
plt.show()

# Heatmap
plt.figure(figsize=(10, 6))
plt.hexbin(df['bill_length_mm'], df['bill_depth_mm'], gridsize=20)
plt.xlabel('Bill Length (mm)')
plt.ylabel('Bill Depth (mm)')
plt.title('Heatmap')
plt.show()

# 3D Scatter Plot
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(df['bill_length_mm'], df['bill_depth_mm'], df['flipper_length_mm'])
ax.set_xlabel('Bill Length (mm)')
ax.set_ylabel('Bill Depth (mm)')
ax.set_zlabel('Flipper Length (mm)')
ax.set_title('3D Scatter Plot')
plt.show()
```
